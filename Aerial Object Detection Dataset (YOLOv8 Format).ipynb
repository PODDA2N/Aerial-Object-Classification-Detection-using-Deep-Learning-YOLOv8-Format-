#### Create data.yaml

```python
import yaml
import os
```

```python
data = {
    "path": r"E:\Data Analyst Coding\Machine Learning Project Internship\YOLOv8 Dataset",       # main dataset folder  
    "train": "images/train",
    "val": "images/val",
    "test": "images/test",
    "names": {
        0: "bird",
        1: "drone"
    }
}

yaml_path = r"E:\Data Analyst Coding\Machine Learning Project Internship\YOLOv8 Dataset\data.yaml"

with open(yaml_path, "w") as f:
    yaml.dump(data, f)

print("data.yaml created at:", yaml_path)
```

#### Train YOLOv8 Model

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(data = r"E:\Data Analyst Coding\Machine Learning Project Internship\YOLOv8 Dataset\data.yaml", epochs=30, imgsz=640)
```

#### Validate Model

```python
metrics = model.val()
print(metrics)
```

#### Streamlit App (YOLOv8)

```python
import streamlit as st
from ultralytics import YOLO
from PIL import Image
import numpy as np

# Load model
@st.cache_resource
def load_model():
    return YOLO(r"E:\Data Analyst Coding\Machine Learning Project Internship\runs\detect")

model = load_model()

st.title("🦅 Bird vs Drone Detection (YOLOv8)")

val_path = r"E:\Data Analyst Coding\Machine Learning Project Internship\runs\detect\val"

if st.button("Run Detection on Validation Images"):

    image_files = os.listdir(val_path)

    if len(image_files) == 0:
        st.error("No images found in val folder ❌")

    for file in image_files:
        if file.endswith((".jpg", ".png", ".jpeg")):

            file_path = os.path.join(val_path, file)

            image = Image.open(file_path)

            # Run YOLO
            results = model.predict(image)

            result_img = results[0].plot()

            st.image(result_img, caption=file)
```
